## Notebook Overview

This notebook explores the relationship between mental health and various demographic, academic, and lifestyle factors in students. The dataset contains information on students' gender, age, academic pressure, study satisfaction, sleep duration, dietary habits, suicidal thoughts, study hours, financial stress, family history of mental illness, and depression.

Using this data, we aim to:
- Analyze how different factors like sleep duration, study satisfaction, and financial stress contribute to students' mental health.
- Investigate the correlation between demographic characteristics (gender, age) and depression.
- Visualize key patterns to understand the prevalence of depression and its relationship with other factors.

In this analysis, we will use Python libraries such as **pandas**, **seaborn**, and **matplotlib** to clean, visualize, and interpret the dataset to provide valuable insights into the mental well-being of students.

The notebook is divided into sections that cover:
- **Data Preprocessing**: Cleaning the dataset and transforming categorical variables.
- **Exploratory Data Analysis (EDA)**: Visualizing relationships and distributions.
- **Advanced Insights**: Correlation analysis and insights into factors contributing to depression in students.

This analysis can be used for identifying mental health risks, promoting preventive strategies, and improving well-being among students.

```python
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('/kaggle/input/depression-student-dataset/Depression Student Dataset.csv')
data.head()
```

```python
# Dataset overview
data.info()
```

```python
data.isnull().sum()
```

## Exploratory Data Analysis (EDA)

```python
# Gender distribution
sns.countplot(x='Gender', data=data, palette='viridis')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

## How does depression vary with gender?

```python
# Depression by gender
sns.countplot(x='Depression', hue='Gender', data=data, palette='coolwarm')
plt.title('Depression Status by Gender')
plt.xlabel('Depression')
plt.ylabel('Count')
plt.show()
```

##  Is there a correlation between academic pressure and depression?

```python
sns.boxplot(x='Depression', y='Academic Pressure', data=data, palette='pastel')
plt.title('Academic Pressure vs Depression')
plt.xlabel('Depression')
plt.ylabel('Academic Pressure')
plt.show()
```

## How does sleep duration affect depression?

```python
# Sleep duration and depression
sns.countplot(x='Sleep Duration', hue='Depression', data=data, palette='muted')
plt.title('Sleep Duration vs Depression')
plt.xlabel('Sleep Duration (hours)')
plt.ylabel('Count')
plt.show()
```

## What is the relationship between study satisfaction and financial stress?

```python
# Study satisfaction vs financial stress
sns.regplot(x='Study Satisfaction', y='Financial Stress', data=data, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Study Satisfaction vs Financial Stress')
plt.xlabel('Study Satisfaction')
plt.ylabel('Financial Stress')
plt.show()
```

```python
data.columns = data.columns.str.strip()  # Remove leading/trailing spaces
print(data.columns)
```

## What is the prevalence of depression among students?

```python
# Depression prevalence
depression_counts = data['Depression'].value_counts()

plt.figure(figsize=(6, 4))
sns.barplot(x=depression_counts.index, y=depression_counts.values, palette='rocket')
plt.title('Depression Prevalence')
plt.xlabel('Depression (Yes/No)')
plt.ylabel('Count')
plt.show()
```
